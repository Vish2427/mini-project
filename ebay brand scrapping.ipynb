```python
df={}
l=[]
s=[]
for i in df1['sku']:
    ebay_url="https://www.ebay.com.au/itm/"+ str(i)
    prodRes= requests.get(ebay_url)
    prodRes.encoding='utf-8'
    prod_html = bs(prodRes.text,'html.parser')
    #print(prod_html)
    try:
        spec=prod_html.find('div',{'data-testid':"ux-layout-section__item", 'class':"ux-layout-section__item ux-layout-section__item--table-view"}).text
    except:
        pass
    t=''
    
    
    if 'Brand:' in spec:
        spec.find('Brand:')
        t=spec[spec.find('Brand:'):spec.find('Brand:')+50]
        a=t[6:]
        if ":" not in a:
            l.append(i)
            s.append(a)
            print(i,a)
        else:
            k=' '.join(a[:a.find(':')].split(' ')[:-1])
            l.append(i)
            s.append(k)
            print(i,k)
        
    
    else:
        pass
df['asin']=l
df['brand']=s
```
